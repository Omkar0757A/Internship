```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://presidentofindia.nic.in/former-presidents.htm')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
name = []

for i in soup.find_all('div', class_="presidentListing"):
    name.append(i.text.split('(')[0].strip())
9    
name
```

```python
term = []

for i in soup.find_all('div', class_="presidentListing"):
    term.append(i.text.split(':')[1].strip().split('\n')[0])

term

```

```python
import pandas as pd
df = pd.DataFrame({'Name of President':name, 'Term of Office':term})
df
```
