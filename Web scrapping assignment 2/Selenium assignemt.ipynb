```python
#Q8
import selenium
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings('ignore')
import time
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
```

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager


driver = webdriver.Chrome(service=Service(ChromeDriverManager(version='114.0.5735.90').install()))
```

```python
driver.get("https://www.azquotes.com/")
```

```python
top_quotes_link = driver.find_element(By.XPATH, '//a[contains(text(), "Top Quotes")]')
top_quotes_link.click()
```

```python
time.sleep(2)
```

```python
quotes_list = []
authors_list = []
types_list = []
```

```python
while len(quotes_list) < 1000:
    quote=driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in quote:
        quotes_list.append(i.text)
        
        if len(quotes_list) == 1000:
            break
         
    authors=driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in authors:
        authors_list.append(i.text)
        
        if len(authors_list) == 1000:
            break
            
    types=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in types:
        types_list.append(i.text)
        
        if len(types_list) == 1000:
            break
    
    try:
        next_button=driver.find_element(By.XPATH,'//li[@class="next"]')
        next_button.click()
        time.sleep(3)
    except:
        break
```

```python
import pandas as pd
df = pd.DataFrame({'Quotes':quotes_list, 'Author Name':authors_list, 'Type':types_list})
df
```

```python
driver.get("https://www.jagranjosh.com/") 
```

```python
gk_option = driver.find_element(By.XPATH, '//a[contains(text(), "GK")]')
gk_option.click()
```

```python
pm_list_option = driver.find_element(By.XPATH, '//a[contains(text(), "List of all Prime Ministers of India")]')
pm_list_option.click()
```

```python
Name=[]
born_dead=[]
term_of_office=[]
remark=[]
```

```python
names = driver.find_elements(By.XPATH,'//div[@class="table-box"]/table/tbody/tr/td[2]/p')

for name in names:

    Name.append(name.text)
bd = driver.find_elements(By.XPATH,'//div[@class="table-box"]/table/tbody/tr/td[3]/p')

for bodo in bd:

    born_dead.append(bodo.text)
tm = driver.find_elements(By.XPATH,'//div[@class="table-box"]/table/tbody/tr/td[4]/p[1]')

for tmo in tm:

    term_of_office.append(tmo.text)
    
rmk = driver.find_elements(By.XPATH,'//div[@class="table-box"]/table/tbody/tr/td[5]/p')

for rmrk in rmk:

    remark.append(rmrk.text)
```

```python
import pandas as pd
df = pd.DataFrame({'Prime-minister name':Name, 'Born-Dead':born_dead,'Term of office':term_of_office, 'Remark':remark})
df
```
